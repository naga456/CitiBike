```python
import numpy as np
import pandas as pd
```

```python
jan = pd.read_csv("./201801-citibike-tripdata.csv")
```

```python
feb = pd.read_csv("./201802-citibike-tripdata.csv")
```

```python
mar = pd.read_csv("./201803-citibike-tripdata.csv")
```

```python
combined = jan
combined = combined.append(feb)
combined = combined.append(mar)
```

```python
#Remove tripduration < 60
combined = combined[combined.tripduration >60]
```

```python
#Remove gender = 0 (unknown)
combined = combined[combined.gender != 0]
```

```python
#Remove birth year < 1919 (age over 100)
combined = combined[combined['birth year'] > 1919]
```

```python
#Add age = 2019 - birth year
combined['age'] = 2019 - combined['birth year']
```

```python
#Modify tripduration to be in minutes
combined['duration'] = combined['tripduration']/60
combined = combined.drop(columns=['tripduration'])
```

```python
#Remove duration > 1440 (There are 1440 minutes in a day)
combined = combined[combined.duration <=1440]
```
